```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Optional: Set a style for Seaborn
sns.set(style="whitegrid")
```

```python
# Load the data from CSV files
consistency_results = pd.read_csv('consistency_results_test_case_1.csv')
similarity_analysis = pd.read_csv('similarity_analysis_test_case_1.csv')
```

```python
# Plot consistency status for each field
plt.figure(figsize=(10, 6))
sns.countplot(data=consistency_results, x='Field', hue='Consistent')
plt.title('Consistency Status by Field')
plt.xlabel('Field')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Consistent')
plt.show()
```

```python
# Merge consistency and similarity data
merged_data = pd.merge(consistency_results[['Field', 'Consistent']], 
                       similarity_analysis.rename(columns={'Unnamed: 0': 'Field'}),
                       on='Field')

# Bar plot with similarity score for consistent vs inconsistent fields
plt.figure(figsize=(12, 6))
sns.barplot(data=merged_data, x='Field', y='Mean Similarity', hue='Consistent')
plt.title('Mean Similarity Score by Field with Consistency Status')
plt.xlabel('Field')
plt.ylabel('Mean Similarity Score')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Consistent')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize DataFrame for similarity scores
data = {
    'Field': ['generalExplanation', 'detailedExplanation', 'generalAdvice', 'tailoredAdvice', 'pharmacologicalExplanation'],
    'Similarity': [1.00, 1.00, 1.00, 1.00, 1.00]  # Based on your description
}
similarity_df = pd.DataFrame(data)
```

```python
# Set style for the plot
sns.set(style="whitegrid")

# Plot similarity scores by field
plt.figure(figsize=(8, 5))
sns.barplot(data=similarity_df, x='Field', y='Similarity')
plt.title('Similarity Scores by Field (Perfect Similarity = 1.0)')
plt.ylim(0, 1)  # Adjust y-axis to focus on scores near 1.00
plt.xlabel('Field')
plt.ylabel('Similarity Score')
plt.xticks(rotation=45, ha='right')
plt.show()
```
